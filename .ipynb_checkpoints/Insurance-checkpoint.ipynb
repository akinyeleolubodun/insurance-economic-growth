### Objective

#### Research Questions
- What is the impact of Life insurance businesses penetration on economic growth in Nigeria?
- What is the impact of Non-Life insurance businesses penetration on economic growth in Nigeria?
- What is the impact of Life insurance businesses density on economic growth in Nigeria?
- What is the impact of Non-Life insurance businesses density on economic growth in Nigeria?

#### Research Hypotheses
- H0 - Life insurance businesses penetration has no significant impact on economic growth in Nigeria.
- H0 - Non-Life insurance businesses penetration has a significant impact on economic growth in Nigeria.
- H0 - Life insurance businesses density has no significant impact on economic growth in Nigeria.
- H0 - Non-Life insurance businesses density has a significant impact on economic growth in Nigeria.

The primary hypothesis seeks to ascertain the collective impact of all independent variables, GDP and Urban Population, which represent economic growth, on the dependent variables, Gross Premium income, reflecting the performance of the insurance sector.

Where, Insurance density represents the average per capita premiums collected, and insurance penetration represent the contribution of the insurance sector to the GDP.

Information about the data shows that apart from the year that is in int datatype, other records are in string which we would not be able to manipulate or get good information from. It will be good to convert the string to a float so we are able to work better with it.

The first step to take is to remove the "," in the number. Then we convert to float.

```python
# Here is function that will help remove CC from the Engine column and also convert the column type to integer


def remove_string_from_number_column(var, str_to_remove):
    if isinstance(var, str):  # checking if the value is string or not
        if (
            str_to_remove in var
        ):  # checking for the presence of str_to_remove in the value
            return var.lower().strip().replace(str_to_remove.lower(), "")
        else:
            return var
    else:
        pass
```

```python
df["Gross Premium Non-Life (USD)"] = df["Gross Premium Non-Life (USD)"].apply(
    remove_string_from_number_column, str_to_remove=","
)
df["Gross Premium Life (USD)"] = df["Gross Premium Life (USD)"].apply(
    remove_string_from_number_column, str_to_remove=","
)
df["GDP (USD)"] = df["GDP (USD)"].apply(
    remove_string_from_number_column, str_to_remove=","
)
df["URBAN POPULATION"] = df["URBAN POPULATION"].apply(
    remove_string_from_number_column, str_to_remove=","
)

df.head()
```

Let us convert to float

```python
# Converting columns to categorical data
columns = {
    "Gross Premium Non-Life (USD)": "float",
    "Gross Premium Life (USD)": "float",
    "GDP (USD)": "float",
    "URBAN POPULATION": "float",
}

df = df.astype(columns)
```

### Test for Stationarity in our Series

```python
from statsmodels.tsa.stattools import adfuller


def adf_test(series):
    result = adfuller(series)
    print("ADF Statistic: %f" % result[0])
    print("p-value: %f" % result[1])
    if result[1] <= 0.05:
        print("The series is stationary")
    else:
        print("The series is not stationary")
```

```python
adf_test(df["GDP (USD)"])
```

### Now, we can proceed to predict the 10-year premium income for both Life and Non-Life insurance. To do this, we can use a time series forecasting model.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming "df" is your DataFrame with cleaned data
X = df["Years"].values.reshape(-1, 1)  # Years as the independent variable
y_life = df["Gross Premium Life (USD)"]
y_non_life = df["Gross Premium Non-Life (USD)"]

# Create a linear regression model for Life insurance
model_life = LinearRegression()
model_life.fit(X, y_life)

# Create a linear regression model for Non-Life insurance
model_non_life = LinearRegression()
model_non_life.fit(X, y_non_life)

# Predict the next 10 years
years_future = np.arange(2021, 2031).reshape(-1, 1)
predicted_life = model_life.predict(years_future)
predicted_non_life = model_non_life.predict(years_future)

# Print the predictions for Life and Non-Life insurance
print("Predicted Life Insurance Premium for the Next 10 Years:")
print(predicted_life)

print("Predicted Non-Life Insurance Premium for the Next 10 Years:")
print(predicted_non_life)
```

### We calculate the percentage change (growth rate) from year to year for both Life and Non-Life insurance premiums and then plot the results.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named "df" with columns "Years" and "Gross Premium Life (USD)"
# You can calculate the growth rate for Life insurance premium as follows:
df["Life Premium Growth Rate (%)"] = df["Gross Premium Life (USD)"].pct_change() * 100

# You can do the same for Non-Life insurance premium:
df["Non-Life Premium Growth Rate (%)"] = (
    df["Gross Premium Non-Life (USD)"].pct_change() * 100
)

# Plot the growth rates
plt.figure(figsize=(12, 6))
plt.plot(
    df["Years"], df["Life Premium Growth Rate (%)"], label="Life Premium Growth Rate"
)
plt.plot(
    df["Years"],
    df["Non-Life Premium Growth Rate (%)"],
    label="Non-Life Premium Growth Rate",
)
plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")
plt.title("Premium Income Growth Rates")
plt.legend()
plt.grid(True)

plt.show()
```

### Below is the contribution of Life and Non-Life insurance premiums to the GDP over the 35-year study period

```python
import pandas as pd

# Assuming you have a DataFrame named "df" with columns "Years", "Gross Premium Life (USD)", and "Gross Premium Non-Life (USD)"
# and "GDP (USD)"

# Calculate the contributions to GDP
df["Life Contribution to GDP (%)"] = (
    df["Gross Premium Life (USD)"] / df["GDP (USD)"]
) * 100
df["Non-Life Contribution to GDP (%)"] = (
    df["Gross Premium Non-Life (USD)"] / df["GDP (USD)"]
) * 100

# Display the results
print(df[["Years", "Life Contribution to GDP (%)", "Non-Life Contribution to GDP (%)"]])
```
